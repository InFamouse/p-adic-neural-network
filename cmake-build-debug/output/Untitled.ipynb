```python
import pandas as pd
from datetime import datetime
import csv
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
def Plot(x, y):
    # plot
    plt.plot(x,y)
    # beautify the x-labels
    plt.gcf().autofmt_xdate()
    plt.show()
```

```python
headers = ['Number of digits','Accuracy', 'Time']
df = pd.read_csv('../../cmake-build-release/output/num_digits.csv',names=headers)
print(df)

# x = df['Number of digits']
# y = df['Accuracy']
Plot(df['Number of digits'], df['Accuracy'])
Plot(df['Number of digits'], df['Time'])
```
