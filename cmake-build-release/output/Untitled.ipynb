```python
def Plot(x, y):
    plt.plot(x, y)
    # beautify the x-labels
#     plt.gcf().autofmt_xdate()
    plt.show()

def PlotScatter(x, y):
    plt.scatter(x, y)
    plt.show()


def PlotSmooth(x, y):
    xnew = np.linspace(x.min(), x.max(), 300) #300 represents number of points to make between T.min and T.max
    tck = interpolate.splrep(x, y)
    smooth = interpolate.splev(xnew, tck)
    plt.plot(xnew, smooth)
    plt.show()
    
    
def PlotSmooth(x, y1, y2, y3):
    xnew = np.linspace(x.min(), x.max(), 300) #300 represents number of points to make between T.min and T.max
    tck = interpolate.splrep(x, y1)
    smooth = interpolate.splev(xnew, tck)
    plt.plot(xnew, smooth, label="train")
    tck = interpolate.splrep(x, y2)
    smooth = interpolate.splev(xnew, tck)
    plt.plot(xnew, smooth, label="valid")
    tck = interpolate.splrep(x, y3)
    smooth = interpolate.splev(xnew, tck)
    plt.plot(xnew, smooth, label="test")
    plt.legend()
    plt.show()
    
def PlotBoolean(n):
    headers = ['Epoch', 'Train accuracy', 'Valid accuracy', 'Test accuracy',  'Time']
    df = pd.read_csv('boolean' + str(n) + '.csv', names=headers)
    # print(df)
    PlotSmooth(df['Epoch'], df['Train accuracy'], df['Valid accuracy'], df['Test accuracy'])
    Plot(df['Epoch'], df['Time'])
    print("Total time:  ", sum(df['Time']))
    
def PlotThree(str):
    headers = ['Epoch', 'Train accuracy', 'Valid accuracy', 'Test accuracy',  'Time']
    df = pd.read_csv(str, names=headers)
    # print(df)
    PlotSmooth(df['Epoch'], df['Train accuracy'], df['Valid accuracy'], df['Test accuracy'])
    Plot(df['Epoch'], df['Time'])
    print("Total time:  ", sum(df['Time']))
    return df
```

```python
df = PlotThree("cancer.csv")
max_count_file = open("max_count.txt","r")
df['Number of subsets'] = np.asarray([int(el) for el in max_count_file.read().split()])
PlotScatter(df['Epoch'], df['Number of subsets'])
```
